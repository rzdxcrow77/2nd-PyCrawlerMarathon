## 1. 取出今天所有的新聞

```python
import requests
from bs4 import BeautifulSoup
from selenium import webdriver


browser = webdriver.Chrome(executable_path='chromedriver')
browser.get("https://www.ettoday.net/news/news-list.htm")

from datetime import datetime
date=datetime.now().strftime("%d")
bBreak=True
import time
while bBreak:
    for i in range(50):
        time.sleep(1)
        browser.execute_script("window.scrollTo(0, 900000);")
    html_source = browser.page_source
    soup = BeautifulSoup(html_source, "html5lib")
    for d in soup.find(class_="part_list_2").find_all('h3'):
        text=d.find(class_="date").text
        print(text, d.find_all('a')[-1].text)
        if text.split('/')[2].split(' ')[0] != date:
            bBreak = False
            break
```

## 2. 取出現在時間兩小時內的新聞

```python
import requests
from bs4 import BeautifulSoup
from selenium import webdriver


browser = webdriver.Chrome(executable_path='chromedriver')
browser.get("https://www.ettoday.net/news/news-list.htm")

from datetime import datetime, timedelta
date=(datetime.now()-timedelta(hours=3)).strftime("%H")
bBreak=True
import time
while bBreak:
    for i in range(25):
        time.sleep(1)
        browser.execute_script("window.scrollTo(0, 900000);")
    html_source = browser.page_source
    soup = BeautifulSoup(html_source, "html5lib")
    for d in soup.find(class_="part_list_2").find_all('h3'):
        text=d.find(class_="date").text
        print(text, d.find_all('a')[-1].text)
        if text.split('/')[2].split(' ')[1].split(':')[0] == date:
            bBreak = False
            break
```

## 3. 根據範例，取出三天前下午三點到五點的新聞

```python
from datetime import datetime, timedelta
date=(datetime.now()-timedelta(days=3)).strftime("%Y-%m-%d")
url='https://www.ettoday.net/news/news-list-{}-0.htm'.format(date)

import requests
from bs4 import BeautifulSoup
from selenium import webdriver


browser = webdriver.Chrome(executable_path='chromedriver')
browser.get(url)

from datetime import datetime, timedelta
bBreak=True
import time
while bBreak:
    for i in range(25):
        time.sleep(1)
        browser.execute_script("window.scrollTo(0, 900000);")
    html_source = browser.page_source
    soup = BeautifulSoup(html_source, "html5lib")
    for d in soup.find(class_="part_list_2").find_all('h3'):
        text=d.find(class_="date").text
        if text.split('/')[2].split(' ')[1].split(':')[0] in ['15','16','17']:
            print(text, d.find_all('a')[-1].text)
        if text.split('/')[2].split(' ')[1].split(':')[0] == '14':
            bBreak = False
            break
```
