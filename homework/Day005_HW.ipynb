```python
# 1. é€™å€‹ API ä¸€æ¬¡æœƒå›å‚³å¹¾ç­†è³‡æ–™ï¼Ÿæ¯ä¸€ç­†è³‡æ–™åŒ…å«å“ªäº›æ¬„ä½ï¼Ÿ

import requests
r = requests.get('https://www.dcard.tw/_api/forums/pet/posts?popular=true')
response = r.text

import json
data = json.loads(response)

n=len(data)
print(n)
```

```python
# 2. å–å‡ºæ¯ä¸€ç­†è³‡æ–™çš„ã€Œæ¨™é¡Œã€ã€ã€Œè²¼æ–‡æ™‚é–“ã€ã€ã€Œç•™è¨€äººæ•¸ã€ã€ã€ŒæŒ‰è®šäººæ•¸ã€

for d in data:
    print("{} , {} , {} , {}".format(d['title'],d['createdAt'],d['commentCount'],d['likeCount']))
    cct+=int(d['commentCount'])
    lct+=int(d['likeCount'])
```

```python
# 3. è¨ˆç®—ç†±é–€/éç†±é–€æ–‡ç« çš„ã€Œå¹³å‡ç•™è¨€äººæ•¸ã€èˆ‡ã€Œå¹³å‡æŒ‰è®šäººæ•¸ã€

r = requests.get('https://www.dcard.tw/_api/forums/pet/posts?popular=false')
response = r.text
data = json.loads(response)

fcct=0
flct=0
fn=len(data)
for d in data:
    fcct+=int(d['commentCount'])
    flct+=int(d['likeCount'])

print("Popular: Comment:{} Like:{}".format(cct/n,lct/n))
print("Non-Popular: Comment:{:6.2f} Like:{:2.2f}".format(fcct/fn,flct/fn))
```
