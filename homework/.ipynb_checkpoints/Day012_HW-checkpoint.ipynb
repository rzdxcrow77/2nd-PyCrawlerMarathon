### ① 取出今天所有的發文

```python
import requests
from bs4 import BeautifulSoup

url = 'https://www.ettoday.net/news/news-list.htm'
r = requests.get(url)

#print(r.text)

soup = BeautifulSoup(r.text, "html5lib")
gps={}
for d in soup.find(class_="part_list_2").find_all('h3'):
    print(d.find(class_="date").text, d.find_all('a')[-1].text,d.find_all('em')[0].text)
    key=d.find_all('em')[0].text
    l=gps.get(key,None)
    if l is None:
        gps.update({key:[(d.find(class_="date").text + d.find_all('a')[-1].text)]})
    else:
        l.append((d.find(class_="date").text + d.find_all('a')[-1].text))
```

### ② 如果想要依照類別分類，怎麼儲存會比較好？

```python
print(list(gps.keys()))
```

```python
maxlen=0
for k,v in gps.items():
    print(k,len(v))
```
